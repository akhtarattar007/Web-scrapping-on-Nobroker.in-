#### Automated the webscrapping for each city url using for loop

```python
demo_name=[]
demo_rate=[]
demo_emi=[]
demo_area=[]
for i in url:
  res = requests.get(url=i)
  html_content = res.content
  soup = BeautifulSoup(html_content,'html.parser')
  locname = soup.find_all('h2') # location of name
  locrate = soup.find_all('div',class_='font-semi-bold heading-6') # location of price,emi,area

  
  #Name
  Name=[]                       
  for j in locname:
    Name.append(j.a.text)      
  demo_name.append(Name)

  #Rate
  rate=[]
  lorate=[]
  for j in locrate:
    rate.append(j.text)
  for k in range(0,len(rate),3):
    lorate.append(rate[k])
  demo_rate.append(lorate)
    
  #Area
  area=[]
  for j in range(2,len(rate),3):
    area.append(rate[j])
  demo_area.append(area)

  #EMI
  EMI=[]
  for j in range(1,len(rate),3):
    EMI.append(rate[j])
  demo_emi.append(EMI)     
```

```python
len(demo_name)
len(demo_rate)
len(demo_emi)
len(demo_area)
```

```python
#BHK
title = []
for i in demo_name:
    for j in i:
        title.append(j)
# print(title)


# Price
Price=[]
for i in demo_rate:
    for j in i:
        Price.append(j)
# print(Price)


# EMI
emi = []
for i in demo_emi:
    for j in i:
        emi.append(j)
# print(emi)


# Area
Area=[]
for i in demo_area:
    for j in i:
        Area.append(j)
# print(Area)
```

```python
df=pd.DataFrame({'Flat_Title':title,'EMI':emi,'Area':Area,'Price':Price})
```
